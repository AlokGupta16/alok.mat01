```python
import matplotlib.pyplot as plt
import pandas as pd

```

```python
df = pd.read_csv("pokemon.csv")
df
```

```python
df.info()
```

```python
plt.plot(df["Generation"],df["Attack"].sort_values(),
color = "green",
linestyle = "--",
marker = "x",
markersize = 8,
label = "Generation",
alpha = 0.8)
plt.legend()

plt.show()
```

```python
plt.scatter(df["Speed"],df["Attack"],
c = "Blue",
s = 100,
marker = "o",
alpha = 0.8)
plt.xlabel("Speed")
plt.ylabel("Attack")

plt.show()
```

```python
plt.bar(df["Type 1"],df["Speed"],
edgecolor = "Blue",
alpha = 0.8)

plt.show()
```

```python
plt.barh(df["Type 1"],df["Speed"],
edgecolor = "Blue",
alpha = 0.8)

plt.show()
```

```python
plt.hist(df["Generation"],
bins = 30,
color = "Blue",
edgecolor = "Black",
density = True)

plt.show()
```

```python
plt.boxplot(df["Speed"],
showmeans = True,
patch_artist = True)

plt.show()
```

```python
plt.pie(df["Type 1"].value_counts(),
labels = df["Type 1"].value_counts().index,
autopct = "%1.1f%%",
startangle = 90)

plt.show()
```

```python
plt.subplot(121),
plt.bar(df["Type 1"],df["Speed"],
            color = "Yellow",
            edgecolor = "Red",
            alpha = 0.8)
plt.xlabel("Type 1")
plt.ylabel("Speed")
plt.xticks(rotation = 90)
plt.title("Speed by Type 1")
plt.tight_layout()

plt.subplot(122),
plt.bar(df["Type 1"],df["Speed"],
            color = "Yellow",
            edgecolor = "Red",
            alpha = 0.8)
plt.xlabel("Type 1")
plt.ylabel("Speed")
plt.xticks(rotation = 90)
plt.title("Speed by Type 1")
plt.tight_layout()
plt.show()
```
